```python
# Qual o volume de pedidos por mês? Existe sazonalidade nas vendas?

ordens = pd.read_csv("06ordem.csv")

# Converter a data para datetime
ordens['order_purchase_timestamp'] = pd.to_datetime(ordens['order_purchase_timestamp'])

# Criar coluna com o período (ano-mês)
ordens['ano_mes'] = ordens['order_purchase_timestamp'].dt.to_period('M')

# Extrair o mês (1 a 12) separadamente
ordens['mes'] = ordens['order_purchase_timestamp'].dt.month

# Contar pedidos por ano-mês
pedidos_por_mes_ano = (
    ordens
    .groupby('ano_mes')
    .size()
    .reset_index(name='qtd')
)

# Adicionar a coluna 'mes' novamente para fazer a média por mês
pedidos_por_mes_ano['mes'] = pedidos_por_mes_ano['ano_mes'].dt.month

# Calcular a média de pedidos por mês (sazonalidade)
media_por_mes = (
    pedidos_por_mes_ano
    .groupby('mes')['qtd']
    .mean()
    .reset_index()
    .rename(columns={'qtd': 'media_pedidos'})
)

# Plotar a média por mês
plt.figure(figsize=(10, 5))
plt.bar(media_por_mes['mes'], media_por_mes['media_pedidos'], color='skyblue')
plt.title("Média de Pedidos por Mês (Sazonalidade)")
plt.xlabel("Mês")
plt.ylabel("Média de Pedidos")
plt.xticks(range(1, 13))
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Mostrar a tabela
print(media_por_mes)
```

```python
# Qual a distribuição do tempo de entrega dos pedidos?

ordens = pd.read_csv("06ordem.csv")

# Converter colunas de data
ordens['order_purchase_timestamp'] = pd.to_datetime(ordens['order_purchase_timestamp'])
ordens['order_delivered_customer_date'] = pd.to_datetime(ordens['order_delivered_customer_date'])

# Calcular o tempo de entrega em dias
ordens['tempo_entrega'] = (ordens['order_delivered_customer_date'] - ordens['order_purchase_timestamp']).dt.days

# Remover valores nulos ou negativos (casos com falha nos dados)
ordens_filtradas = ordens[ordens['tempo_entrega'].notna() & (ordens['tempo_entrega'] >= 0)]

# Plotar histograma da distribuição
plt.figure(figsize=(10, 5))
plt.hist(ordens_filtradas['tempo_entrega'], bins=30, color='mediumseagreen', edgecolor='black')
plt.title("Distribuição do Tempo de Entrega dos Pedidos")
plt.xlabel("Tempo de Entrega (dias)")
plt.ylabel("Quantidade de Pedidos")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Estatísticas básicas
print(ordens_filtradas['tempo_entrega'].describe())
```

```python
# Qual a relação entre o valor do frete e a distância de entrega?


def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Raio da Terra em km
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(dlambda/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

# Carregar dados
itens = pd.read_csv("03itens.csv")
ordens = pd.read_csv("06ordem.csv")
clientes = pd.read_csv("01clientes.csv")
vendedores = pd.read_csv("08vendedores.csv")
geolocalizacao = pd.read_csv("02localização.csv")

# Remover duplicatas na localização
geolocalizacao = geolocalizacao.groupby('geolocation_zip_code_prefix')[['geolocation_lat', 'geolocation_lng']].mean().reset_index()

# Juntar dados
itens_ordens = pd.merge(itens, ordens[['order_id', 'customer_id']], on='order_id', how='left')
itens_ordens = pd.merge(itens_ordens, clientes, left_on='customer_id', right_on='id_cliente', how='left')
itens_ordens = pd.merge(itens_ordens, vendedores, on='seller_id', how='left')

# Juntar coordenadas do cliente e do vendedor
itens_ordens = pd.merge(itens_ordens, geolocalizacao, left_on='cep_prefixo', right_on='geolocation_zip_code_prefix', how='left')
itens_ordens = pd.merge(itens_ordens, geolocalizacao, left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix', suffixes=('_cliente', '_vendedor'), how='left')

# Calcular distância
itens_ordens['distancia_km'] = haversine(
    itens_ordens['geolocation_lat_cliente'],
    itens_ordens['geolocation_lng_cliente'],
    itens_ordens['geolocation_lat_vendedor'],
    itens_ordens['geolocation_lng_vendedor']
)

# Filtrar valores válidos
df_valido = itens_ordens[itens_ordens['distancia_km'].notna() & (itens_ordens['freight_value'] > 0)]

# Plotar gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(df_valido['distancia_km'], df_valido['freight_value'], alpha=0.3, color='royalblue')
plt.xlabel('Distância (km)')
plt.ylabel('Valor do Frete (R$)')
plt.title('Relação entre Distância e Valor do Frete')
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlação
correlacao = df_valido[['distancia_km', 'freight_value']].corr().iloc[0, 1]
print(f"Correlação entre distância e frete: {correlacao:.2f}")
```

```python
# Quais são as categorias de produtos mais vendidas em termos de faturamento?


itens = pd.read_csv("03itens.csv")
produtos = pd.read_csv("07produtos.csv")

# Juntar produtos com itens (para ter a categoria)
itens_produtos = pd.merge(itens, produtos, on='product_id', how='left')

# Calcular faturamento por categoria
faturamento_categoria = (
    itens_produtos
    .groupby('product_category_name')['price']
    .sum()
    .reset_index()
    .rename(columns={'price': 'faturamento'})
    .sort_values(by='faturamento', ascending=False)
)

# Exibir as 10 principais categorias
top_categorias = faturamento_categoria.head(10)

# Plotar
plt.figure(figsize=(12, 6))
plt.barh(top_categorias['product_category_name'], top_categorias['faturamento'], color='teal')
plt.xlabel("Faturamento (R$)")
plt.title("Top 10 Categorias de Produtos por Faturamento")
plt.gca().invert_yaxis()  # Categoria mais alta no topo
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Mostrar tabela
print(top_categorias)
```

```python
# Quais estados brasileiros possuem o maior valor médio de pedido?


ordens = pd.read_csv("06ordem.csv")
itens = pd.read_csv("03itens.csv")
clientes = pd.read_csv("01clientes.csv")

# Juntar ordens com clientes para obter o estado
ordens_clientes = pd.merge(ordens[['order_id', 'customer_id']], clientes, left_on='customer_id', right_on='id_cliente', how='left')

# Juntar com itens para obter os valores dos pedidos
dados = pd.merge(itens[['order_id', 'price']], ordens_clientes, on='order_id', how='left')

# Calcular valor total por pedido
total_pedidos = dados.groupby(['order_id', 'estado'])['price'].sum().reset_index(name='valor_pedido')

# Calcular valor médio por estado
valor_medio_estado = (
    total_pedidos
    .groupby('estado')['valor_pedido']
    .mean()
    .reset_index()
    .sort_values(by='valor_pedido', ascending=False)
)

# Exibir resultado
print("Valor médio dos pedidos por estado:")
print(valor_medio_estado)

# Visualizar em gráfico
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(valor_medio_estado['estado'], valor_medio_estado['valor_pedido'], color='coral')
plt.title('Valor Médio dos Pedidos por Estado')
plt.xlabel('Estado')
plt.ylabel('Valor Médio (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Calcule a taxa de clientes recorrentes. Considere um cliente recorrente aquele que fez mais de um pedido no período analisado. 
# Quais insights podemos extrair destes dados?
ordens = pd.read_csv("06ordem.csv")

# Contar número de pedidos por cliente
pedidos_por_cliente = ordens.groupby('customer_id')['order_id'].nunique().reset_index(name='num_pedidos')

# Classificar cliente como recorrente (mais de 1 pedido)
pedidos_por_cliente['recorrente'] = pedidos_por_cliente['num_pedidos'] > 1

# Calcular total de clientes e quantos são recorrentes
total_clientes = pedidos_por_cliente.shape[0]
recorrentes = pedidos_por_cliente['recorrente'].sum()
taxa_recorrencia = recorrentes / total_clientes * 100

# Mostrar resultado
print(f"Total de clientes únicos: {total_clientes}")
print(f"Clientes recorrentes: {recorrentes}")
print(f"Taxa de recorrência: {taxa_recorrencia:.2f}%")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Carregar dados
ordens = pd.read_csv("06ordem.csv")

# Remover registros com datas ausentes
ordens = ordens.dropna(subset=[
    'order_delivered_customer_date', 
    'order_estimated_delivery_date',
    'order_purchase_timestamp'
])

# Converter datas
ordens['order_delivered_customer_date'] = pd.to_datetime(ordens['order_delivered_customer_date'])
ordens['order_estimated_delivery_date'] = pd.to_datetime(ordens['order_estimated_delivery_date'])
ordens['order_purchase_timestamp'] = pd.to_datetime(ordens['order_purchase_timestamp'])

# Definir variável alvo: atraso (1 = atrasado, 0 = no prazo)
ordens['atrasado'] = (ordens['order_delivered_customer_date'] > ordens['order_estimated_delivery_date']).astype(int)

# Criar features (você pode adicionar mais se quiser)
ordens['tempo_estimado_entrega'] = (ordens['order_estimated_delivery_date'] - ordens['order_purchase_timestamp']).dt.days
ordens['dia_semana_compra'] = ordens['order_purchase_timestamp'].dt.weekday
ordens['mes_compra'] = ordens['order_purchase_timestamp'].dt.month

# Features e alvo
X = ordens[['tempo_estimado_entrega', 'dia_semana_compra', 'mes_compra']]
y = ordens['atrasado']

# Dividir treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo Random Forest
modelo = RandomForestClassifier(random_state=42)
modelo.fit(X_train, y_train)

# Previsões
y_pred = modelo.predict(X_test)

# Avaliação
print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred))
```

```python
# Crie visualizações e dashboards que respondam às seguintes necessidades:

# Um dashboard geral que mostre a evolução das vendas ao longo do tempo, com filtros por estado e categoria de produto

# Um mapa de calor mostrando a concentração de vendas por região/estado do Brasil

# Um conjunto de gráficos que apresente a relação entre avaliação do cliente e tempo de entrega

# Um dashboard de análise dos vendedores, mostrando quais têm melhor desempenho em termos de volume de vendas, satisfação do cliente e tempo de entrega
ordens = pd.read_csv("06ordem.csv", parse_dates=['order_purchase_timestamp'])
itens = pd.read_csv("03itens.csv")
clientes = pd.read_csv("01clientes.csv")
produtos = pd.read_csv("07produtos.csv")
categorias = pd.read_csv("09produto_categoria.csv")
revisoes = pd.read_csv("05revisao_pedidos.csv")

# Unificar dados
dados = itens.merge(ordens[['order_id', 'customer_id', 'order_purchase_timestamp',
                            'order_delivered_customer_date', 'order_estimated_delivery_date']],
                    on='order_id', how='left') \
             .merge(clientes[['id_cliente', 'estado']], left_on='customer_id', right_on='id_cliente', how='left') \
             .merge(produtos[['product_id', 'product_category_name']], on='product_id', how='left') \
             .merge(categorias, on='product_category_name', how='left') \
             .merge(revisoes[['order_id', 'review_score']], on='order_id', how='left')

# Calcular métricas
dados['faturamento'] = dados['price'] + dados['freight_value']
dados['ano_mes'] = dados['order_purchase_timestamp'].dt.to_period('M')
dados['tempo_entrega'] = (
    pd.to_datetime(dados['order_delivered_customer_date']) -
    pd.to_datetime(dados['order_purchase_timestamp'])
).dt.days

# Gráfico 1: Evolução das vendas
plt.figure(figsize=(12, 6))
vendas = dados.groupby('ano_mes')['faturamento'].sum()
vendas.plot(marker='o')
plt.title('Evolução das Vendas ao Longo do Tempo')
plt.xlabel('Ano-Mês')
plt.ylabel('Faturamento')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico 2: Faturamento por Estado
plt.figure(figsize=(12, 5))
estado_vendas = dados.groupby('estado')['faturamento'].sum().sort_values(ascending=False)
sns.barplot(x=estado_vendas.index, y=estado_vendas.values, palette='coolwarm')
plt.title('Faturamento por Estado')
plt.ylabel('Faturamento')
plt.xlabel('Estado')
plt.tight_layout()
plt.show()

# Gráfico 3: Avaliação x Tempo de entrega
plt.figure(figsize=(10, 5))
avaliacao = dados.groupby('review_score')['tempo_entrega'].mean().reset_index()
sns.barplot(data=avaliacao, x='review_score', y='tempo_entrega', palette='magma')
plt.title('Tempo Médio de Entrega por Avaliação')
plt.xlabel('Nota de Avaliação')
plt.ylabel('Tempo Médio de Entrega (dias)')
plt.tight_layout()
plt.show()

# Gráfico 4: Desempenho de Vendedores
vendedores = dados.groupby('seller_id').agg({
    'order_id': 'count',
    'faturamento': 'sum',
    'tempo_entrega': 'mean',
    'review_score': 'mean'
}).reset_index().rename(columns={
    'order_id': 'qtd_vendas',
    'faturamento': 'total_vendas',
    'tempo_entrega': 'media_entrega',
    'review_score': 'satisfacao'
})

top_vendedores = vendedores.sort_values(by='total_vendas', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_vendedores, x='seller_id', y='total_vendas', palette='viridis')
plt.title('Top 10 Vendedores por Faturamento')
plt.ylabel('Faturamento')
plt.xlabel('ID do Vendedor')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
